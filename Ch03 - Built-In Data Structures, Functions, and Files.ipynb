````python
def apply_to_list(some_list, f): 
    return [f(x) for x in some_list] 

ints = [4, 0, 1, 5, 6] 
apply_to_list(ints, lambda x: x * 2) 
````

Great question — and your intuition is correct: **there *is* iteration happening**, but in the second case **Python is doing it internally** for you.

---

## ✅ What happens in `apply_to_list`

You explicitly wrote the loop:

```python
[f(x) for x in some_list]
```

So you can clearly see:

* iterate over each element
* call `f(x)`
* collect results

---

## ✅ What happens in `strings.sort(key=...)`

Here:

```python
strings.sort(key=lambda x: len(set(x)))
```

You didn’t write the loop — but **`sort()` does it internally**.

### Internally, Python does something like this:

For each string in the list:

1. Take the element `x`
2. Compute `key(x)`
3. Use that key value to decide sorting order

So effectively, Python is doing something like:

```python
for x in strings:
    k = len(set(x))
```

But it doesn’t return that list of keys — it uses the keys to sort.

---

## 🧠 The key idea: `key=` is a *function applied to each element*

The `key=` argument expects a function.

Python’s sorting algorithm will call it like:

```python
key(strings[0])
key(strings[1])
key(strings[2])
...
```

So the iteration is hidden inside `.sort()`.

---

## 🔥 Example: What keys get computed?

```python
strings = ["foo", "card", "bar", "aaaa", "abab"]
```

Let’s compute the key for each:

| string | set(string)       | unique letters | key = len(set(x)) |
| ------ | ----------------- | -------------- | ----------------- |
| "foo"  | {'f','o'}         | 2              | 2                 |
| "card" | {'c','a','r','d'} | 4              | 4                 |
| "bar"  | {'b','a','r'}     | 3              | 3                 |
| "aaaa" | {'a'}             | 1              | 1                 |
| "abab" | {'a','b'}         | 2              | 2                 |

So Python sorts by these key values:

```python
["aaaa", "foo", "abab", "bar", "card"]
```

---

## ✅ Important detail: key is computed once per element (not repeatedly)

Python uses an optimization called the **Schwartzian transform** (decorate-sort-undecorate):

Internally it does something like:

```python
decorated = [(key(x), x) for x in strings]
decorated.sort()
strings = [x for (k, x) in decorated]
```

So your lambda isn’t called 1000 times during comparisons — it’s usually called **once per element**, then those key values are used.

##### Slicing

````python
seq = [7, 2, 3, 7, 5, 6, 0, 1]
seq[1:5]
````

````python
seq[:5]
````

````python
seq[3:]
````

Negative indices slice the sequence relative to the end:

````python
seq[-4:]
````

````python
seq[-6:-2]
````

````python
seq
````

````python
seq[::-1]
````

### Functions

Since Python functions are objects, many constructs can be easily expressed that are difficult to do in other languages. Suppose we were doing some data cleaning and needed to apply a bunch of transformations to the following list of strings:

````python
states = ["   Alabama ", "Georgia!", "Georgia", "georgia", "FlOrIda", "south   carolina##", "West virginia?"]
````

````python
import re
def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub("[!#?]", "", value)
        value = value.title()
        result.append(value)
    return result
````

````python
clean_strings(states)
````

alternate approach

````python
def remove_punctuation(value):
    return re.sub("[!#?]", "", value)
clean_ops = [str.strip, remove_punctuation, str.title]
def clean_strings(strings, ops):
    result = []
    for value in strings:
        for func in ops:
            value = func(value)
        result.append(value)
    return result

clean_strings(states, clean_ops)
````

A generator is a convenient way, similar to writing a normal function, to construct a new iterable object. Whereas normal functions execute and return a single result at a time, generators can return a sequence of multiple values by pausing and resuming execution each time the generator is used. To create a generator, use the yield keyword instead of return in a function:

````python
def squares(n=10):
    print(f"Generating squares from 1 to {n ** 2}")
    for i in range(1, n + 1):
        yield i ** 2
````

````python
gen = squares()
````

````python
gen
````

````python
for x in gen:
    print(x, end=" ")
````

gen becomesn iterator.

Great question — this is one of Python’s nicest “you don’t have to think about it” features.

### Why does the file close automatically?

Because `with ... as ...` uses something called a **context manager**.

When you write:

```python
with open(path, encoding="utf-8") as f:
    lines = [x.rstrip() for x in f]
```

Python does this behind the scenes:

1. **Opens the file**
2. Gives you the file object as `f`
3. Runs the block inside `with`
4. **No matter what happens**, it closes the file at the end

Even if:

* an exception happens inside the block
* you return early from a function
* your loop breaks
* anything crashes inside that block

The file still gets closed.

---

### What Python *really* expands it into

This is roughly equivalent to:

```python
f = open(path, encoding="utf-8")
try:
    lines = [x.rstrip() for x in f]
finally:
    f.close()
```

That `finally:` is the key.

`finally` always runs — success or failure.

---

### How does `with` know what to call?

The file object returned by `open()` implements two special methods:

* `__enter__()` → called when entering the `with`
* `__exit__()` → called when leaving the `with`

So conceptually:

```python
f = open(...)
f.__enter__()
# run block
f.__exit__()
```

And inside `__exit__()`, Python calls `close()` for you.

---

### Summary in one line

✅ `with open(...) as f:` ensures `f.close()` happens automatically, even if errors occur.
