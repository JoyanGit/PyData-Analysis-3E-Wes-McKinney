```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

@interact(lam=(0.1,5,0.1), a=(0,5,0.2), b=(0.5,8,0.2))
def plot_exp_prob(lam=1, a=1, b=3):

    x = np.linspace(0,10,400)
    y = lam*np.exp(-lam*x)

    plt.figure(figsize=(7,4))
    plt.plot(x,y)

    mask = (x>=a)&(x<=b)
    plt.fill_between(x[mask],y[mask],alpha=0.3)

    prob = np.exp(-lam*a)-np.exp(-lam*b)

    plt.title(f'Probability = {prob:.4f}')
    plt.grid()
    plt.show()
```
